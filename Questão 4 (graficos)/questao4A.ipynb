```python
# ALUNO: RAFAEL SCHMIDT DAL PAZ

import numpy as np
import matplotlib.pyplot as plt

# Definindo a faixa de valores de -3 a 5
w = np.linspace(-3, 5, 100)

# z(w):
z_w = 3 / (3 + 1j * w)

# Calculando a magnitude|z(w)|
magnitude = np.abs(z_w)

# Plot do gráfico magnitude
plt.figure(figsize=(8, 5))
plt.plot(w, magnitude, label=r'$|Z(w)| = \left| \frac{3}{3 + jw} \right|$', color='b')
plt.axhline(y=1, color='gray', linestyle='--', linewidth=0.7)  # Linha guia em 1
plt.xlabel(r'$\omega$', fontsize=12)
plt.ylabel(r'$|Z(w)|$', fontsize=12)
plt.title('Gráfico da Magnitude de $Z(w)$', fontsize=14)
plt.legend()
plt.grid()
plt.show()


# Calculando a fase
fase = np.angle(z_w)

# Plot do gráfico fase
plt.figure(figsize=(8, 5))
plt.plot(w, fase, label=r'$\angle Z(w) = \tan^{-1}(-w/3)$', color='r')
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.7)  # Linha guia em 0
plt.xlabel(r'$\omega$', fontsize=12)
plt.ylabel(r'Fase $\theta(w)$ (rad)', fontsize=12)
plt.title('Gráfico da Fase de $Z(w)$', fontsize=14)
plt.legend()
plt.grid()
plt.show()
```
