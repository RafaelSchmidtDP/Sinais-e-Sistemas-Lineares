```python
# ALUNO: RAFAEL SCHMIDT DAL PAZ

import numpy as np
import matplotlib.pyplot as plt
import math

def magnitude_4b(w):
    #Calcula a magnitude do número complexo Z2(ω) da questão 4B.
    return 5 / np.sqrt(9 + w**2)

def fase_4b(w):
    #Calcula a fase do número complexo Z2(ω) em graus.
    if w > 0:
        return 90 - math.degrees(math.atan(w / 3))
    elif w < 0:
        return -90 - math.degrees(math.atan(w / 3))
    else:
        return None

w = np.linspace(-10, 10, 500)  # Cria um vetor de frequências angulares

# Calcula magnitude e fase para cada valor de w
magnitude_valores_4b = magnitude_4b(w)
fase_valores_4b = [fase_4b(val) for val in w]  # Calcula a fase para cada valor de w

# Remove os valores None (para w=0) para plotar
w_plot = [val for val, f in zip(w, fase_valores_4b) if f is not None]
fase_valores_plot = [f for f in fase_valores_4b if f is not None]

# Cria os gráficos
plt.figure(figsize=(12, 6))

# Gráfico da Magnitude (Questão 4B)
plt.subplot(2, 1, 1)
plt.plot(w, magnitude_valores_4b, label=r'$|Z_2(w)| = \frac{5}{\sqrt{9 + w^2}}$', color='b')
plt.axhline(y=5/3, color='gray', linestyle='--', linewidth=0.7)  # Linha guia em 5/3 (magnitude em w=0)
plt.xlabel(r'$\omega$', fontsize=12)
plt.ylabel(r'$|Z_2(w)|$', fontsize=12)
plt.title('Gráfico da Magnitude de $Z_2(w)$ (Questão 4B)', fontsize=14)
plt.legend()
plt.grid(True)

# Gráfico da Fase (Questão 4B)
plt.subplot(2, 1, 2)
plt.plot(w_plot, fase_valores_plot, label=r'$\angle Z_2(w)$', color='r')
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.7)  
plt.xlabel(r'$\omega$', fontsize=12)
plt.ylabel(r'Fase $\theta_2(w)$ (graus)', fontsize=12)
plt.title('Gráfico da Fase de $Z_2(w)$ (Questão 4B)', fontsize=14)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
